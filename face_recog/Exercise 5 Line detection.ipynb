```python
### code from https://www.learnopencv.com/hough-transform-with-opencv-c-python/ modified for quite good results from
### random picture from internet
### Exercise 5

import cv2
import numpy as np
import matplotlib.pyplot as plt

# Read image 
img = cv2.imread('data/road4.jpeg') # road.png is the filename
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2GRAY))
plt.show()

sub_img = img[int(0.6*img.shape[0]):img.shape[0],0:img.shape[1]]
plt.imshow(cv2.cvtColor(sub_img, cv2.COLOR_BGR2GRAY))
plt.show()

# Convert the image to gray-scale
gray = cv2.cvtColor(sub_img, cv2.COLOR_BGR2GRAY)
# Find the edges in the image using canny detector
gausian = cv2.GaussianBlur(gray, (19, 19), 0)
edges = cv2.Canny(gausian, 70, 60)

#cv2 imshow crashes jupyter, use plt.imshow instead
#cv2.imshow("Result Image", edges)
#cv2.waitKey(0)
plt.imshow(edges)
plt.show()

# Detect points that form a line
lines = cv2.HoughLinesP(edges, 1, np.pi/180, threshold=70, minLineLength=60, maxLineGap=175)
print(lines)

# Draw lines on the image
for line in lines:
    x1, y1, x2, y2 = line[0]
    cv2.line(sub_img, (x1, y1), (x2, y2), (255, 0, 0), 3)
# Show result
#cv2.imshow("Result Image", sub_img)
#cv2.waitKey(0)
#cv2.destroyAllWindows()
plt.imshow(sub_img)
plt.show()
```
